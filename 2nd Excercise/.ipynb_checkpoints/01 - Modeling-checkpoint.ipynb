```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

import tensorflow as tf
from tensorflow.keras.layers import Normalization, Dense, InputLayer
from tensorflow.keras.losses import MeanSquaredError, Huber, MeanAbsoluteError
from tensorflow.keras.metrics import RootMeanSquaredError

from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.metrics import confusion_matrix, classification_report

import math

import joblib
```

```python
df = pd.read_csv("pancreatic_cancer_prediction.csv")
```

```python
df_0 = df[df['Survival_Status'] == 0].sample(5000)
df_1 = df[df['Survival_Status'] == 1]
df = pd.concat([df_0, df_1])
```

```python
target = 'Survival_Status'
cat_column = []
num_column = []
two_val = []
to_drop = []
for col in df.columns:
    if col not in to_drop and col != target:
        if df[col].nunique() == 2:
            two_val.append(col)
        elif df[col].dtype == 'O':
            cat_column.append(col)
        else:
            num_column.append(col)
```

```python
def drop_columns(X):
    return X.drop(columns=to_drop)
```

```python
# Create subplots
rows = math.ceil(len(cat_column)/2)
fig, axes = plt.subplots(nrows=rows, ncols=2, figsize=(15, 10*rows))
axes = axes.flatten()

for i, col in enumerate(cat_column):
    sns.countplot(data=df, x=col, hue=target, ax=axes[i])
    axes[i].set_title(f'{col} vs {target}')
    axes[i].legend(title=target, loc='upper right')

# Remove the last empty subplot if the number of columns is odd
if len(cat_column) % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
### Imput
num_pipeline = Pipeline([
    ('imput',SimpleImputer(strategy='mean')),
    ('scaling', StandardScaler())
])
cat_pipeline = Pipeline([
    ('imput', SimpleImputer(strategy='most_frequent')),
    ('encoding', OneHotEncoder())
])
two_val_pipeline = Pipeline([
    ('imput', SimpleImputer(strategy='most_frequent')),
    ('encoding', OrdinalEncoder())
])
```

```python
preprocessor = ColumnTransformer([
    ('numerical', num_pipeline, num_column),
    ('categorical', cat_pipeline, cat_column),
    ('ordinal', two_val_pipeline, two_val)
])
pipepline = Pipeline([
    ('drop', FunctionTransformer(drop_columns)),
    ('prepro', preprocessor)
])
```

```python
joblib.dump(pipepline, 'prepprocessor.pkl')
```

```python
x = df.drop(columns=target)
y = df[target]
xs_train, xs_test, ys_train, ys_test = train_test_split(x, y, test_size=0.2, random_state=44)
```

```python
svm_model = Pipeline([
    ('drop_column', FunctionTransformer(drop_columns)),
    # ('drop_na', FunctionTransformer(drop_na)),
    ('preprocessing', preprocessor),
    ('model', SVC(kernel='sigmoid'))
])
```

```python
svm_model
```

```python
svm_model.fit(xs_train, ys_train)
```

```python
y_predict_svm = svm_model.predict(xs_test)
```

```python
confusion_matrix(ys_test, y_predict_svm)
```

```python
df[target].value_counts()
```

```python
print(classification_report(ys_test, y_predict_svm))
```

```python
X_transformed = pipepline.fit_transform(x)
```

```python
X_transformed
```

```python
xd_train, xd_test, yd_train, yd_test = train_test_split(X_transformed, y, test_size=0.2, random_state=44)

# Define the neural network model
```

```python
xd_train.shape
```

```python
input_shape = xd_train.shape[1]
```

```python
input_shape
```

```python
  # Determine input shape dynamically
deep_model = tf.keras.Sequential([
    InputLayer(input_shape=(input_shape,)),
    Dense(128, activation="relu"),
    Dense(64, activation="relu"),
    Dense(32, activation="relu"),
    Dense(2, activation='sigmoid')
])
```

```python
deep_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# model.fit(X_train, y_train, epochs=150, validation_data=(X_test, y_test))
deep_model.fit(xd_train, yd_train, epochs=50, validation_data=(xd_test, yd_test))
```

```python
deep_model.save('deep_model.keras')
```

```python
# Make predictions on the test data
y_pred = deep_model.predict(xd_test)
y_pred[0:, 0]>0.5

# Convert the probabilities to binary class predictions (0 or 1)
y_pred_class = y_pred.argmax(axis=1)  # Takes the class with the highest probability (0 or 1)

# Now you can generate and display the confusion matrix
cm = confusion_matrix(yd_test, y_pred_class)

cm
```
