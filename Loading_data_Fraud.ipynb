```python
import joblib
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('Dataset/card_transdata.csv')
```

```python
x = df.drop(columns='fraud')
```

```python
def prediction(x):
    excluded_columns  = []
    x = x.drop(columns = excluded_columns)
    model = tf.keras.models.load_model('credit_card_fraud.h5')
    preprocessor = joblib.load('credit_card_fraud.pkl')
    t= preprocessor.fit_transform(x)
    y_pred = model.predict(t)
    y_pred = np.argmax(y_pred, axis=1)
    y_pred = pd.DataFrame({'Prediction':y_pred})
    y_pred = y_pred.replace({1:'No',0:'Yes'})
    return pd.concat([x.reset_index(),y_pred], axis = 1)
```

```python
prediction(x)
```
